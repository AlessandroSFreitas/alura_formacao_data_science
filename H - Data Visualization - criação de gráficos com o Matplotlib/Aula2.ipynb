#### Importando as bibliotecas

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Importando os dados

```python
df = pd.read_csv("monitoramento_tempo.csv")
```

#### Leitura dos dados

```python
df.head()
```

```python
# transformando o atributo "data" da tabela para um atributo do tipo DateTime.

df["data"] = pd.to_datetime(df["data"])
```

#### Customizando nossas visualizações

```python
# determinando eixos

x = "temperatura"
y = "data"
```

```python
# utilizando o parâmetro lw para manipular a espessura da linha.

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(y, x, data=df, color="g", lw=0.4) # <------------------
eixo.set_title("Temperatura no momento", fontdict={"fontsize": 18})
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["Temperatura"], loc="lower right", fontsize=14)
```

```python
# utilizando o parâmetro ls para mudar o estilo da linha. 

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(y, x, data=df, color="g", ls="dotted") # <------------------
eixo.set_title("Temperatura no momento", fontdict={"fontsize": 18})
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["Temperatura"], loc="lower right", fontsize=14)
```

```python
# limitando a visualização entre um certo período, no eixo X.

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(y, x, data=df, color="g")
eixo.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1)) # <-----------------
eixo.set_title("Temperatura no momento", fontdict={"fontsize": 18})
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["Temperatura"], loc="lower right", fontsize=14)
```

```python
# utilizando o parâmetro marker para alterar o estilo da linha.

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(y, x, data=df, color="g", marker="o") # <---------------
eixo.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1))
eixo.set_title("Temperatura no momento", fontdict={"fontsize": 18})
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["Temperatura"], loc="lower right", fontsize=14)
```

```python
# adicionando um grid na visualização.

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.plot(y, x, data=df, color="g")
eixo.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1))
eixo.set_title("Temperatura no momento", fontdict={"fontsize": 18})
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["Temperatura"], loc="lower right", fontsize=14)
eixo.grid(True) # <-----------------------
```

```python
# criando as figuras
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

# configurando a primeira figura
eixo.grid(True)
eixo.plot(y, x, data=df, color="g")
eixo.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1))
eixo.set_title("Temperatura em 2014", fontdict={"fontsize": 25}, pad=15)
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["temperatura"], loc="lower right", fontsize=15)

# configurando a segunda figura, a figura de dentro
eixo2.grid(True)
eixo2.plot(y, x, data=df, color="b")
eixo2.set_title("Temperatura no momento", fontdict={"fontsize": 15})
eixo2.set_ylabel("Temperatura", fontdict={"fontsize": 11})
eixo2.set_xlabel("Data", fontdict={"fontsize": 11})
```

```python
# criando as figuras
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

# configurando a primeira figura
eixo.grid(True)
eixo.plot(y, x, data=df, color="g")
eixo.set_xlim(datetime(2014, 5, 1), datetime(2014, 6, 1))
eixo.set_ylim(270, 320)
eixo.set_title("Temperatura em Maio/2014", fontdict={"fontsize": 25}, pad=15)
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["temperatura"], loc="lower right", fontsize=15)

# adicionando filtros
azul_esquerda = df["data"] < datetime(2014, 5, 1)
azul_direita = df["data"] > datetime(2014, 6, 1)

# configurando a segunda figura, a figura de dentro
eixo2.grid(True)
eixo2.plot(y, x, data=df, color="g")
eixo2.plot(df[azul_esquerda]["data"], df[azul_esquerda]["temperatura"], color="b")
eixo2.plot(df[azul_direita]["data"], df[azul_direita]["temperatura"], color="b")
eixo2.set_xlim(datetime(2014, 1, 1), datetime(2015, 1, 1))
eixo2.set_title("Temperatura em 2014", fontdict={"fontsize": 15})
eixo2.set_ylabel("Temperatura", fontdict={"fontsize": 11})
eixo2.set_xlabel("Data", fontdict={"fontsize": 11})
```

#### Analisando Temperatura máxima/mínima da série histórica

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.grid(True)
eixo.plot(y, x, data=df, color="g")
eixo.set_title("Temperatura no momento", fontdict={"fontsize": 25})
eixo.set_xlabel("Data", fontdict={"fontsize": 20})
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 20})
eixo.legend(["Temperatura"], loc="lower right", fontsize=15)

x1_max = df["data"][df["temperatura"].idxmax()]
y1_max = max(df["temperatura"])

x2_max = df["data"][df["temperatura"].idxmax() - 7000]
y2_max = max(df["temperatura"] - 5)

eixo.annotate("Máximo", xy=(x1_max, y1_max), fontsize=20, xytext=(x2_max, y2_max), arrowprops=dict(facecolor="k"))

x1_min = df["data"][df["temperatura"].idxmin()]
y1_min = min(df["temperatura"])

x2_min = df["data"][df["temperatura"].idxmin() - 7000]
y2_min = min(df["temperatura"] + 5)

eixo.annotate("Mínimo", xy=(x1_min, y1_min), fontsize=20, xytext=(x2_min, y2_min), arrowprops=dict(facecolor="k"))

# adicionando linha nas temperaturas máximas e mínimas
eixo.axhline(max(df["temperatura"]), color="k", linestyle="--")
eixo.axhline(min(df["temperatura"]), color="k", linestyle="--")
```

```python
temperatura_por_dia_da_semana = df.groupby("dia_da_semana")["temperatura"].mean()
nome_dias = ["Domingo", "Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado"]
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0, 0, 1, 1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(nome_dias, temperatura_por_dia_da_semana, color=cores, edgecolor='black')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
         autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', size=15, pad=10)
```
