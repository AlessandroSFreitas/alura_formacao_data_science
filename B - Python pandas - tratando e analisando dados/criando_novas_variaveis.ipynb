## Criando novas variáveis

```python
import pandas as pd
```

```python
url = 'https://drive.google.com/file/d/1bf38agrPvyHiPS855ECIU0UR9QzXG4_o/view?usp=sharing'

file_id = url.split('/')[-2]

download_url = 'https://drive.google.com/uc?id=' + file_id

dados = pd.read_csv(download_url, sep=';')
```

```python
dados.head()
```

```python
# Criar novo atributo para o DataFrame com a soma total de Valor, Condominio, IPTU -> Valor Bruto

dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
# Criar novo atributo para o DataFrame com a soma de Valor e Area -> Valor m2

dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
# Deixa apenas 2 (duas) casa depois da virgula, no atributo Valor m2

dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head(10)
```

```python
# Cria um novo atributo para o DataFrame com a divisão entre Valor Bruto e Area -> Valor Bruto m2 e por cima, deixa apenas
# 2 (duas) casa depois da virgula

dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# Cria um novo atributo, aplicando Casa no atributo Tipo Agregado caso sejam dos tipos Casa, Casa de Condomínio, Casa de Vila

dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(100)
```
