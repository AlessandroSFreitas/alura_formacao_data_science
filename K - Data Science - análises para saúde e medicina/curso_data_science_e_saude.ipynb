```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")

recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
```

```python
display(deaths.head())
display(recovered.head())
```

```python
def latest_by_country(data):
    """Get by country the last date update data."""
    return data.groupby("Country/Region").sum().iloc[:, -1]
```

```python
latest_by_country(confirmed)
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
```

```python
combined = pd.concat(informations, axis=1)
combined.columns = ["confirmed", "deaths", "recovered"]
```

```python
combined
```

```python
sum_up = combined.sum()
```

```python
# global letality rate until 8/3/22

letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```

```python
# global letality rate ultil 8/5/22
# https://www.worldometers.info/coronavirus/

deaths_1 = 6431309
confirmed_1 = 587054779

letality = deaths_1 / confirmed_1
print(letality * 100)
```

```python
# letality rate per country

letality_rate_per_country = combined["deaths"] / combined["confirmed"] * 100

combined["letality_rate_per_country"] = letality_rate_per_country
combined.head()
```

```python
# lets see the letality on Brazil

combined.loc["Brazil"]
```

```python
# letality rate per country (head 20 - from minor to major)

combined.sort_values("letality_rate_per_country", ascending=False).head(20)
```

```python
# confirmed per country (head 20 - from minor to major)

combined.sort_values("confirmed", ascending=False).head(20)
```

```python
# deaths per country (head 20 - from minor to major)

combined.sort_values("deaths", ascending=False).head(20)
```

```python
def latest_by_country_at(data, date):
    """Get by country on the specific date"""
    return data.groupby("Country/Region").sum()[date]
```

```python
informations = [
    latest_by_country_at(confirmed, "2/20/20"), 
    latest_by_country_at(deaths, "2/20/20"), 
    latest_by_country_at(recovered, "2/20/20")
]

combined_2_20_20 = pd.concat(informations, axis=1)
combined_2_20_20.columns = ["confirmed", "deaths", "recovered"]

sum_up = combined_2_20_20.sum()

# global letality rate until 2/20/20

letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```

```python
informations = [
    latest_by_country_at(confirmed, "2/8/20"), 
    latest_by_country_at(deaths, "2/20/20"), 
    latest_by_country_at(recovered, "2/20/20")
]

combined_12 = pd.concat(informations, axis=1)
combined_12.columns = ["confirmed", "deaths", "recovered"]

sum_up = combined_12.sum()

# global letality rate until 2/20/20

letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```
