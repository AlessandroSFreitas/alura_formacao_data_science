#### Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Importando os dados

```python
df = pd.read_csv("monitoramento_tempo.csv")
```

#### Leitura dos dados

```python
df.head()
```

```python
df.info()
```

```python
# transformando o atributo "data" da tabela para um atributo do tipo DateTime.

df["data"] = pd.to_datetime(df["data"])
```

#### Plotando gráficos

```python
plt.figure(figsize=(15, 8))
x = "temperatura"
y = "data"
plt.plot(y, x, data=df)
plt.title("Temperatura no tempo", fontdict={"fontsize": 16})
```

- O atributo "temperatura" não é Celsius, e sim Kelvin.

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
x = "temperatura"
y = "data"
eixo.plot(y, x, data=df, color='green')
eixo.set_title("Temperatura no momento", fontdict={"fontsize": 18})
eixo.set_ylabel("Temperatura", fontdict={"fontsize": 16})
eixo.set_xlabel("Data", fontdict={"fontsize": 16})
eixo.legend(["Temperatura"], loc="lower right", fontsize=14)
```
